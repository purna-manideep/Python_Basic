```python
# 2. Print the maximum length word of given words
# Hai hello world
# hello
# world
words=input().split()
length=[]
for word in words:
    length.append(len(word))
max_length=max(length)
for i in words:
    if max_length==len(i):
        print(i)
```

```python
3. Print all the words which starts with the given character, If no words matches with the given condition you should print -1
#ramu ravi balu

s= input().split()
r=input()
for j in s:
    if j.startswith(r):
        print(j)
print(-1)
```

```python
# 4. Print all the words which words not ends with vowels(a,e,i,o,u)
# ramu ravi balu raju mahesh

s= input().split()
r=['a','e','i','o','u']
for j in s:
    for i in r:
        if j.endswith(i):
            continue
print(j)
```
